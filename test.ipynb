```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import geom

# Создаем данные
x = np.arange(geom.ppf(0.95, 0.5))  # ppf - Percent Point Function
rv = geom(0.5)
plt.bar(x, rv.pmf(x), width=0.8)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Создаем данные
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
count, bins, ignored = plt.hist(s, 30, density=True, alpha=0.7, color='g')

# Теперь добавляем функцию распределения
y = norm.pdf(bins, mu, sigma)
lw = 2
plt.plot(bins, y, 'k', linewidth=lw)
plt.title("Функция плотности вероятности")
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация 1000 случайных чисел из равномерного распределения между 0 и 1
data = np.random.uniform(0,1,1000)

# Построение гистограммы
count, bins, ignored = plt.hist(data, 20, facecolor='green')

# Добавление меток к осям и заголовков графика
plt.xlabel('X~U[0,1]')
plt.ylabel('Count')
plt.title("Гистограмма равномерного распределения (размер ячейки 20)")

# Отображение графика
plt.show()
```
